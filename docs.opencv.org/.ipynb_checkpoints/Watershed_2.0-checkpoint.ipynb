```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img = cv.imread('Y10.jpg')
plt.imshow(img)
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
ret, thresh = cv.threshold(gray,0,255,cv.THRESH_BINARY_INV+cv.THRESH_OTSU)
```

```python
# Remove Salt and Pepper noise
median_filter = cv.medianBlur(gray, 5)
plt.imshow(median_filter, 'gray')
```

```python
#Edge
# 3*3 Sobel Filters
Gx= np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
print ("Gx \n", Gx)
Gy = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]])
print ("Gy \n", Gy)
img_sobelx = cv.Sobel(median_filter,cv.CV_8U,1,0,ksize=3)
img_sobely = cv.Sobel(median_filter,cv.CV_8U,0,1,ksize=3)
#del f = Gx + Gy
# Adding mask to the image
img_sobel = img_sobelx + img_sobely + gray
plt.imshow(img_sobel,cmap='gray')
plt.show()
```

```python
# Threshold.
# take the mean of the pixel values and use that as a threshold.
gray_r = gray.reshape(gray.shape[0] * gray.shape[1])
threshold = gray_r.mean()
maxValue = 255
 
# Threshold the pixel values
th, thresh = cv.threshold(img_sobel, threshold, maxValue, cv.THRESH_BINARY)
plt.imshow(thresh,cmap='gray')
plt.show()
```

```python
# remove any small white noises in the image using morphological opening. 
kernel = np.ones((3,3),np.uint8)
opening = cv.morphologyEx(thresh,cv.MORPH_OPEN,kernel, iterations = 2)

plt.imshow(opening,cmap='gray')
plt.show()
```

```python
# Black region shows sure background area
# Dilation increases object boundary to background.
sure_bg = cv.dilate(opening,kernel,iterations=3)

plt.imshow(sure_bg,cmap='gray')
plt.show()
```

```python
# White region shows sure foreground area
import cv2
dist_transform = cv.distanceTransform(opening, cv2.cv2.DIST_L2,5)
ret, sure_fg = cv.threshold(dist_transform,0.7*dist_transform.max(),255,0)

plt.imshow(sure_fg,cmap='gray')
plt.show()
```

## Watershed Algorithmto to identify unknown regions

```python
sure_fg = np.uint8(sure_fg)
unknown = cv2.subtract(sure_bg,sure_fg)

plt.imshow(unknown,cmap='gray')
plt.show()
```

```python
#Watershed Segmentation
contours, hierarchy = cv.findContours(sure_fg,cv.RETR_TREE,cv.CHAIN_APPROX_SIMPLE)

# Creating a numpy array for markers and converting the image to 32 bit using dtype paramter
marker = np.zeros((gray.shape[0], gray.shape[1]),dtype = np.int32)

marker = np.int32(sure_fg) + np.int32(sure_bg)

# Marker Labelling
for id in range(len(contours)):
    cv.drawContours(marker,contours,id,id+2, -1)

marker = marker + 1

marker[unknown==255] = 0

copy_img = img.copy()

cv2.watershed(copy_img, marker)

imgplt = plt.imshow(marker)
plt.colorbar()
plt.show()
```

```python
copy_img[marker==-1]=(0,0,255)
cv2.imwrite('img.jpg',copy_img)
plt.imshow(copy_img, 'gray')
plt.show()
```

```python
kernel=cv2.getStructuringElement(cv2.MORPH_RECT,(7,7))
erosion = cv2.morphologyEx(median_filter, cv2.MORPH_ERODE, kernel)
plt.imshow(erosion,cmap='gray')
plt.show()
```

```python
# part of the image having more intensity which has the tumor
dilation = cv.morphologyEx(erosion, cv.MORPH_DILATE, kernel)
plt.imshow(dilation,cmap='gray')
plt.axis('off')
plt.show()
```
